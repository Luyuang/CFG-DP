# Offline Eval

```python
import matplotlib.pyplot as plt
from lerobot.common.datasets.lerobot_dataset import LeRobotDataset
import os

# 定义目标文件夹路径
home_path = os.path.expanduser("~")
root = os.path.join(home_path, "luyuang/nls/v0/lerobot")
parts = root.split("/")[-3:]
repo_id = '_'.join(parts[:2]) + '/' + parts[2]
local_files_only = True

# 加载数据集
dataset = LeRobotDataset(repo_id=repo_id, local_files_only=local_files_only, root=root)
num_episodes = dataset.episode_data_index["from"].shape[0]

# 存储每个 episode 的长度
episode_lengths = []

# 遍历所有 episode
for episode_index in range(num_episodes):
    # 获取当前 episode 的起始和结束索引
    from_idx = dataset.episode_data_index["from"][episode_index].item()
    to_idx = dataset.episode_data_index["to"][episode_index].item()
    
    # 计算当前 episode 的长度
    episode_length = to_idx - from_idx
    episode_lengths.append(episode_length)
    print(f"Episode {episode_index}: Length = {episode_length}")

# 绘制分布图
plt.figure(figsize=(10, 6))
plt.hist(episode_lengths, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title("Distribution of Episode Lengths", fontsize=16)
plt.xlabel("Episode Length", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
